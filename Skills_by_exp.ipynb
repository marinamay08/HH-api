```python
import requests
import pandas as pd
import time
import re
```

```python
phrase_to_search = 'NAME:("PHP разработчик")'
```

```python
ses = requests.Session()
ses.headers = {'HH-User-Agent': "Mozilla/5.0 (X11; Linux x86_64; rv:10.0) Gecko/20100101 Firefox/10.0"}

url = f'https://api.hh.ru/vacancies?text={phrase_to_search}&per_page=100'  #здесь может быть включена индустрия, не забудь
res = ses.get(url)

# getting a list of all pesponses
res_all = []
for p in range(res.json()['pages']):
    print(f'scraping page {p}')
    url_p = url + f'&page={p}'
    res = ses.get(url_p)
    res_all.append(res.json())
```

```python
# for every experience with skills
# parcing vacancies ids, getting vacancy page and scraping tags from each vacancy
tags_list = []
experience_list = []
data_list = []

for page_res_json in res_all:
    for item in page_res_json['items']:
        vac_id = item['id']
        vac_res = ses.get(f'https://api.hh.ru/vacancies/{vac_id}')
        if len(vac_res.json()["key_skills"]) > 0:
            print(vac_id)
            experience = vac_res.json()["experience"]["name"]
            print(experience)
            experience_list.append(experience)
            data = vac_res.json()["experience"]
            print()
```

```python
# parcing vacancies ids, getting vacancy page and scraping tags from each vacancy\n",

tags_list_NO = []
experience_list_NO = []

tags_list_1to3 = []
experience_list_1to3 = []

tags_list_3to6 = []
experience_list_3to6 = []

tags_list_more6 = []
experience_list_more6 = []

for page_res_json in res_all:
    for item in page_res_json['items']:
        vac_id = item['id']
        vac_res = ses.get(f'https://api.hh.ru/vacancies/{vac_id}')
        if len(vac_res.json()["key_skills"]) > 0 and vac_res.json()["experience"]["name"] == "Нет опыта":
            print(vac_id)
            tags_NO = [v for v_dict in vac_res.json()["key_skills"] for _, v in v_dict.items()]
            print(' '.join(tags_NO))
            tags_list_NO.append(tags_NO)
            experience_NO = vac_res.json()["experience"]["name"]
            print(experience_NO)
            experience_list_NO.append(experience_NO)
            print()
                    
        elif len(vac_res.json()["key_skills"]) > 0 and vac_res.json()["experience"]["name"] == "От 1 года до 3 лет":
                print(vac_id)
                tags_1to3 = [v for v_dict in vac_res.json()["key_skills"] for _, v in v_dict.items()]
                print(' '.join(tags_1to3))
                tags_list_1to3.append(tags_1to3)
                experience_1to3 = vac_res.json()["experience"]["name"]
                print(experience_1to3)
                experience_list_1to3.append(experience_1to3)
                print()
                
        elif len(vac_res.json()["key_skills"]) > 0 and vac_res.json()["experience"]["name"] == "От 3 до 6 лет":
                print(vac_id)
                tags_3to6 = [v for v_dict in vac_res.json()["key_skills"] for _, v in v_dict.items()]
                print(' '.join(tags_3to6))
                tags_list_3to6.append(tags_3to6)
                experience_3to6 = vac_res.json()["experience"]["name"]
                print(experience_3to6)
                experience_list_3to6.append(experience_3to6)
                print()
                
        elif len(vac_res.json()["key_skills"]) > 0 and vac_res.json()["experience"]["name"] == "Более 6 лет":
                print(vac_id)
                tags_more6 = [v for v_dict in vac_res.json()["key_skills"] for _, v in v_dict.items()]
                print(' '.join(tags_more6))
                tags_list_more6.append(tags_more6)
                experience_more6 = vac_res.json()["experience"]["name"]
                print(experience_more6)
                experience_list_more6.append(experience_more6)
                print()
```

```python
res_NO = {'phrase': phrase_to_search, 'items_number': len(tags_list_NO), 'skills': tags_list_NO, 'experience': experience_list_NO}
res_NO_df = pd.DataFrame(res_NO)
res_NO_df.shape
```

```python
tag_str_NO = str(tags_list_NO)

skills_for_noExperience = pd.Series(''.join(tag_str_NO).lower()
        .replace('[', '')
        .replace(']', '')
        .replace("'", '')
        .replace("'", '')
        .split(', '))\
        .value_counts()

pd_skills_for_noExperience = pd.DataFrame(skills_for_noExperience, columns=['frequency'])
```

```python
res_1to3 = {'phrase': phrase_to_search, 'items_number': len(tags_list_1to3), 'skills': tags_list_1to3, 'experience': experience_list_1to3}

res_1to3_df = pd.DataFrame(res_1to3)

res_1to3_df.shape
```

```python
tag_str_1to3 = str(tags_list_1to3)

skills_for_between1And3 = pd.Series(''.join(tag_str_1to3).lower()\
         .replace('[', '')\
        .replace(']', '')\
        .replace("'", '')\
        .replace("'", '')\
        .split(', '))\
        .value_counts()
                 
pd_skills_for_between1And3 = pd.DataFrame(skills_for_between1And3, columns=['frequency'])
```

```python
res_3to6 = {'phrase': phrase_to_search, 'items_number': len(tags_list_3to6), 'skills': tags_list_3to6, 'experience': experience_list_3to6}

res_3to6_df = pd.DataFrame(res_3to6)

res_3to6_df.shape
```

```python
tag_str_3to6 = str(tags_list_3to6)

skills_for_between3And6 = pd.Series(''.join(tag_str_3to6).lower()\
         .replace('[', '')\
        .replace(']', '')\
        .replace("'", '')\
        .replace("'", '')\
        .split(', '))\
        .value_counts()
                 
pd_skills_for_between3And6 = pd.DataFrame(skills_for_between3And6, columns=['frequency'])
```

```python
res_more6 = {'phrase': phrase_to_search, 'items_number': len(tags_list_more6), 'skills': tags_list_more6, 'experience': experience_list_more6}

res_more6_df = pd.DataFrame(res_more6)
res_more6_df.shape
```

```python
tag_str_more6 = str(tags_list_more6)

skills_for_more6 = pd.Series(''.join(tag_str_more6).lower()\
        .replace('[', '')
        .replace(']', '')
        .replace("'", '')\
        .replace("'", '')\
        .split(', '))\
        .value_counts()
                 
pd_skills_for_more6 = pd.DataFrame(skills_for_more6, columns=['frequency'])
```

```python
pd_skills_for_noExperience['exp'] = 'Нет опыта'
pd_skills_for_between1And3['exp'] = 'От 1 года до 3 лет'
pd_skills_for_between3And6['exp'] = 'От 3 до 6 лет'
pd_skills_for_more6['exp'] = 'Более 6 лет'
```

```python
result_skills = pd.concat([pd_skills_for_noExperience, pd_skills_for_between1And3, pd_skills_for_between3And6, pd_skills_for_more6]).reset_index() 
```

```python
result_skills_pivot = result_skills.rename(columns={'index': 'skills'})\
        .pivot(index='skills', columns='exp', values='frequency')\
        .fillna(0)
```

```python
result_skills_pivot['frequency_sum'] = result_skills_pivot['Более 6 лет'] + result_skills_pivot['Нет опыта'] + result_skills_pivot['От 1 года до 3 лет'] + result_skills_pivot['От 3 до 6 лет']
```

```python
result_skills_pivot.sort_values('frequency_sum', ascending=False)
```

```python
result_skills_pivot.to_csv('skills_for_{}.csv'.format(phrase_to_search))
```
